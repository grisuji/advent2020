```python
d = "e se sw w nw ne"
d = d.split()
def read_line(l):
    ll = l.strip()
    line = []
    while ll:
        if ll[0] in d:
            line.append(ll[0])
            ll = ll[1:]
        else:
            line.append(ll[:2])
            ll = ll[2:]
    return line
        
def read_manu(filename):
    
    with open(filename, 'r') as f:
        result = []
        for l in f:
            result.append(read_line(l))
    return result
            
    
```

```python
class Tile:
    room = {}
    
    def __init__(self, point):
        self.color = "W"
        self.point = point
        self.room[point] = self
        self.tmp = None
    
    def pos(self):
        return self.point
    
    def clean(self):
        self.room = {}
    
    def move(self, d):
        #print("move {}".format(d))
        dd = None
        if d == "e":
            pp = (self.point[0], self.point[1]+2)
        elif d == "se":
            pp = (self.point[0]-1, self.point[1]+1)
        elif d == "sw":
            pp = (self.point[0]-1, self.point[1]-1)
        elif d == "w":
            pp = (self.point[0], self.point[1]-2)
        elif d == "nw":
            pp = (self.point[0]+1, self.point[1]-1)
        elif d == "ne":
            pp = (self.point[0]+1, self.point[1]+1)
        else:
            print("error, wrong direction")

        if pp in self.room:
            dd = self.room[pp]
        else:
            dd = Tile(pp)

        return dd

    def neighbours(self):
        pp = [(self.point[0], self.point[1]+2),
              (self.point[0], self.point[1]-2),
              (self.point[0]-1, self.point[1]+1),
              (self.point[0]-1, self.point[1]-1),
              (self.point[0]+1, self.point[1]-1),
              (self.point[0]+1, self.point[1]+1)]

        return pp
    
    def flip(self):
        self.color = "B" if self.color=="W" else "W"
        if self.color == "B":
            self.fill()
            
        
    def get_color(self):
        return self.color
    
    def bn(self):
        b = 0
        
        for p in self.neighbours():
            if p in self.room:
                if self.room[p].get_color()=="B":
                    b+=1
        return b
        
    def flip2(self):
        b = self.bn()
        if self.color == "B" and (b==0 or b > 2):
            self.tmp = "W"
        if self.color == "W" and b==2:
            self.tmp = "B"
        return self.tmp is not None
            
    def execute(self):
        if self.tmp is not None:
            self.color = self.tmp
            if self.color == "B":
                self.fill()
            self.tmp = None
    
    def fill(self):
        for p in self.neighbours():
            if p not in self.room:
                c = Tile(p)

        
        
def play(l):
    if (0,0) in Tile.room:
        cur = Tile.room[(0,0)]
    else:
        cur = Tile((0,0))
    for m in l:
        cur = cur.move(m)
    cur.flip()
    return cur
        
    
def count_b():
    b=0
    for p, t in Tile.room.items():
        if t.get_color() == "B":
            b += 1
    return b

def day():
    mem = []
    for p, t in Tile.room.items():
        if t.flip2():
            mem.append(t)

    for t in mem:
        t.execute()

```
