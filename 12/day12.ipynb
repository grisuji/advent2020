```python
import math
def read_data(file):
    d = []
    with open(file,'r') as f:
        d = [ (l[0], int(l[1:-1])) for l in f ]
    return d
tdata = read_data("t.txt")
idata = read_data("i.txt")
```

```python
def m(cmd, l):
    if cmd == "S" or cmd == "W":
        l = -l
    if cmd == "N" or cmd == "S":
        return (0, l)
    else:
        return (l, 0)
    
def rotate(wp, direction, deg):
    if direction == "L":
        deg = -deg
    angle = math.radians(deg)
    s = math.sin(angle) 
    c = math.cos(angle) 
    xnew = wp[0] * c + wp[1] * s
    ynew = -wp[0] * s + wp[1] * c
    return (round(xnew), round(ynew))
    
def add_pos(p1, p2):
    return (p1[0]+p2[0], p1[1]+p2[1])

def move(act, position, wp, v=0):
    d = act[0]
    l = act[1]
    p = list(position)
    
    if d == "F":
        position = add_pos(position, (wp[0]*l, wp[1]*l))
    elif d in ["N", "S", "W", "E"]:
        if v==0:
            position = add_pos(position, m(d, l))
        else:
            wp = add_pos(wp, m(d, l))
    elif d in ["R", "L"]:
        wp = rotate(wp, d, l)
    return position, wp
```

```python
position=(0,0)
wp=(1,0)
for mm in idata:
    position, wp = move(mm, position, wp)
    #print("{}  {}".format(mm, position))
abs(position[0])+abs(position[1])
```
