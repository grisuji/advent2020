```python

def transform(subject_no, loop_size):
    div = 20201227
    value=1
    for _ in range(loop_size):
        value *= subject_no
        value %= div
    return value
        
def handshake(card_loopsize, door_loopsize):
    card_pub_key = transform(7, card_loopsize)
    door_pub_key = transform(7, door_loopsize)
    card_encryption_key = transform(door_pub_key, card_loopsize)
    door_encryption_key = transform(card_pub_key, door_loopsize)
    return card_encryption_key, card_encryption_key == door_encryption_key

def guess_loop_size(subject_no, pub_key):
    div = 20201227
    value=1    
    for loop_size in range(1_000_000_000):
        value *= subject_no
        value %= div
        if value == pub_key:
            return loop_size+1
    print("Error")
    return -1
```

```python
card_loopsize = guess_loop_size(7, 5764801)
door_loopsize = guess_loop_size(7, 17807724)
card_loopsize, door_loopsize
```

```python
handshake(card_loopsize, door_loopsize)
```
