```python
def read_data(file):
    d = [0]
    with open(file,'r') as f:
        for l in f:
            d.append(int(l))
    d.append(max(d)+3)
    return d
tdata = read_data("t.txt")
idata = read_data("i.txt")
sdata = read_data("s.txt")
```

```python
tdata.sort()
idata.sort()
sdata.sort()
```

```python
def get_dist(data):
    dist = {1:0, 2:0, 3:0}
    for i in range(len(data)-1):
        #print("{} {}  {}".format(i, data[i+1], data[i]))
        delta = data[i+1]-data[i]
        dist[delta] += 1
    return dist
```

```python
d=get_dist(tdata)
d[1]*d[3]
```

```python
d=get_dist(idata)
d[1]*d[3]
```

```python
def traverse(data,i=0):
    count = 0
    lmax = len(data)
    if i == lmax-1:
        return 1
    for j in range(1,4):
        if i+j < lmax:
            if data[i+j]-data[i] <= 3:
                count += traverse(data, i+j)
        else:
            break
    return count
```

```python
traverse(idata)
```
