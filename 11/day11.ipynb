```python
def read_data(file):
    d = []
    with open(file,'r') as f:
        d = [ l[:-1] for l in f ]
    return d
tdata = read_data("t.txt")
idata = read_data("i.txt")
```

```python
def get_seat(seats, i, j, default):
    if i < 0 or i > len(seats)-1:
        return default
    row = seats[i]
    if j < 0 or j > len(row)-1:
        return default
    return row[j]
    
def is_seat(seats, i, j):
    s = get_seat(seats, i,j, '.')
    return s != '.'

def is_blocked(seats, i, j):
    s = get_seat(seats, i,j, '.')
    return s == '#'

def all_empty(seats, i, j):
    res = []
    for ii in range(-1, 2):
        for jj in range(-1, 2):
            if i!=0 or j!=0:
                res.append(not is_blocked(seats,i+ii,j+jj))
    return all(res)
    
def blocked(seats, i, j):
    res = 0
    for ii in range(-1, 2):
        for jj in range(-1, 2):
            if (ii!=0 or jj!=0) and is_blocked(seats, i+ii,j+jj):
                #print("{}{}{}".format(i+ii,j+jj,is_blocked(seats, i+ii,j+jj)))
                res+=1
    return res

def change(seats, i, j, c):
    row = seats[i]
    s = list(row)
    s[j] = c
    seats[i] = ''.join(s)
    return seats
    
def iter(seats):
    copy = seats[:]
    for i, row in enumerate(seats):
        for j, seat in enumerate(row):
            if is_seat(seats,i,j) and all_empty(seats, i,j):
                copy = change(copy,i,j,"#")
            elif is_blocked(seats, i, j) and blocked(seats, i,j) >=4:
                copy = change(copy,i,j,"L")
    return copy

                
def changed(s1,s2):
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            return True
    return False

def count(seats):
    c = 0
    for i in range(len(seats)):
        c += seats[i].count("#")
    return c

s=tdata
for i in range(1000):
    s1=iter(s)
    if not changed(s,s1):
        break
    s = s1
count(s)
```

```python
def get_seat(seats, i, j, default):
    if i < 0 or i > len(seats)-1:
        return default
    row = seats[i]
    if j < 0 or j > len(row)-1:
        return default
    return row[j]
    
def is_seat(seats, i, j):
    s = get_seat(seats, i,j, '.')
    return s != '.'

def is_inside(seats, i, j):
    s = get_seat(seats, i,j, 'x')
    return s != 'x'

def is_blocked(seats, i, j):
    s = get_seat(seats, i,j, '.')
    return s == '#'

def is_free(seats, i, j):
    s = get_seat(seats, i,j, '.')
    return s == 'L'

def see_blocked(seats, i, j, di, dj):
    for c in range(1,1000):
        if not is_inside(seats, i+c*di, j+c*dj):
            return False
        if is_blocked(seats, i+c*di, j+c*dj):
            return True
        if is_free(seats, i+c*di, j+c*dj):
            return False
    return False

def all_empty(seats, i, j):
    res = []
    for ii in range(-1, 2):
        for jj in range(-1, 2):
            if i!=0 or j!=0:
                res.append(not is_blocked(seats,i+ii,j+jj))
    return all(res)
    
def see_all_empty(seats, i, j):
    res = []
    for ii in range(-1, 2):
        for jj in range(-1, 2):
            if i!=0 or j!=0:
                res.append(not see_blocked(seats,i, j, ii, jj))
    return all(res)
    
def blocked(seats, i, j):
    res = 0
    for ii in range(-1, 2):
        for jj in range(-1, 2):
            if (ii!=0 or jj!=0) and see_blocked(seats, i, j, ii, jj):
                #print("{}{}{}".format(i+ii,j+jj,is_blocked(seats, i+ii,j+jj)))
                res+=1
    return res

    

def change(seats, i, j, c):
    row = seats[i]
    s = list(row)
    s[j] = c
    seats[i] = ''.join(s)
    return seats
    
def iter(seats):
    copy = seats[:]
    for i, row in enumerate(seats):
        for j, seat in enumerate(row):
            if is_seat(seats,i,j) and see_all_empty(seats, i,j):
                copy = change(copy,i,j,"#")
            elif is_blocked(seats, i, j) and blocked(seats, i,j) >=5:
                copy = change(copy,i,j,"L")
    return copy

                
def changed(s1,s2):
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            return True
    return False

def count(seats):
    c = 0
    for i in range(len(seats)):
        c += seats[i].count("#")
    return c

s=tdata
for i in range(1000):
    s1=iter(s)
    if not changed(s,s1):
        break
    s = s1
count(s)
```
