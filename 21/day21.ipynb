```python
from copy import deepcopy

def _read(file):
    r = []
    i = set() # ingredients
    a = set() # allergens
    with open(file, 'r') as f:
        for l in f:
            ll = l.strip()
            ll = ll.split("(")
            ingre = ll[0].split()
            aller = ll[1][:-1]
            aller = aller[9:].replace(",", "")
            aller = aller.split()
            r.append((set(ingre), set(aller)))
            i = i.union(set(ingre))
            a = a.union(set(aller))
    return r,i,a
```

```python
trules, _, _ = _read("t.txt")
irules, _, _ = _read("i.txt")
```

```python
def make_aller2ingres(rules):
    a2i = {}
    ingres = set()

    for ii, aa in rules:
        for aller in aa:
            a2i[aller] = ii.intersection(a2i.get(aller, ii))
            ingres.union(ii)
    return a2i, ingres

def clean_aller2ingres(a2i):
    tmp_a2i = {}
    uniques = set()
    non_uniques = 0
    for aller, ingres in a2i.items(): 
        if len(ingres) == 1:
            uniques.add(next(iter(ingres)))
        else:
            non_uniques += 1
    for aller, ingres in a2i.items(): 
        tmp_a2i[aller] = ingres
        if len(ingres) > 1:
            tmp_a2i[aller].difference_update(uniques)
    
    return tmp_a2i, non_uniques==0

    
def solve(rules):
    a2i, ii = make_aller2ingres(rules)
    for i in range(100):
        a2i, finish = clean_aller2ingres(a2i)
        if finish:
            break
    else:
        print("infinity loop")
    allergen_ingres  = set()
    for ii in list(a2i.values()):
        allergen_ingres.add(next(iter(ii)))
    part1=0    
    for ii, aa in rules:
        for i in ii:
            if i not in allergen_ingres:
                part1 +=1
    
    part2 = ''
    for key in sorted(a2i):
        ingre=next(iter(a2i[key]))
        if part2:
            part2 = part2 + ','
        part2 += ingre
    
    return a2i, allergen_ingres,part1, part2
```

```python
solve(trules)
```

```python
solve(irules)
```
