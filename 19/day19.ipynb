```python
def rread(filename):
    rules = {}
    msg = []
    with open(filename,'r') as f:
        for l in f:
            ll = l.strip()
            if not ll:
                continue
            if ll[0].isdigit():
                r_number, r = ll.split(":")
                r = [ rr.split()  for rr in r.split("|")]
                if r[0][0]=='"a"':
                    rules[r_number] = ['a']
                elif r[0][0]=='"b"':
                    rules[r_number] = ['b']
                else:    
                    rules[r_number] = r
            else:
                msg.append(ll)
    return rules, msg
trules, tmsg = rread("t.txt")
irules, imsg = rread("i.txt")
ttrules, ttmsg = rread("tt.txt")
irules2, imsg2 = rread("i2.txt")
ttrules2, ttmsg2 = rread("tt2.txt")
```

```python
def E(msg, rule, rules):
    result = None
    for e in rule:
        if (e == "a" or e == "b"):
            if msg[0] == e:
                return msg[1:], True
            else:
                return msg[1:], False
        sub_result = True
        rest = msg
        for r in e:
            rest, valid = E(rest, rules[r], rules)
            sub_result = sub_result and valid
            if not sub_result:
                break
        if sub_result:
            return rest, True
    return rest, False

def start(msg, rules):
    return E(msg, rules["0"], rules)
            
```

```python
import copy
ttrules2=copy.deepcopy(ttrules)
ttrules2["8"].append(['42', '8'])
ttrules2["11"].append(['42', '11', '31'])

irules2=copy.deepcopy(irules)
irules2["8"].append(['42', '8'])
irules2["11"].append(['42', '11', '31'])
```

```python
def E(msg, rule_no, rules, level):
    if not msg:
        return set(), False
    resultset = set()
    result = False
    for rule in rules[rule_no]:  # Bsp:[['8'], ['8', '42']]
        if (rule == "a" or rule == "b"):
            if msg[0] == rule:
                #print("{}{}".format(stack,msg[0]))
                return set([msg[1:]]), True
            else:
                return set([msg]), False
            
        rest = set([msg])
        for subrule in rule: # Bsp:['8', '42']
            subresultset = set() # keep only the last
            for subrest in rest:
                _rest, match = E(subrest, subrule, rules, level+1)   
                if match:
                    subresultset = subresultset.union(_rest) 
                    result = True
            rest = subresultset.copy()
            if not result:
                break
        resultset = resultset.union(subresultset)
    return resultset, result

def start(msg, rules):
    return E( msg, "0", rules, level=0)
```
