```python
def rread(filename):
    rules = {}
    msg = []
    with open(filename,'r') as f:
        for l in f:
            ll = l.strip()
            if not ll:
                continue
            if ll[0].isdigit():
                r_number, r = ll.split(":")
                r = [ rr.split()  for rr in r.split("|")]
                if r[0][0]=='"a"':
                    rules[r_number] = ['a']
                elif r[0][0]=='"b"':
                    rules[r_number] = ['b']
                else:    
                    rules[r_number] = r
            else:
                msg.append(ll)
    return rules, msg
trules, tmsg = rread("t.txt")
irules, imsg = rread("i.txt")
ttrules, ttmsg = rread("tt.txt")
```

```python
def E(pmsg, msg, rule, rules, level, rn=0):
    if len(msg) == 0:
        return msg, False
    for e in rule:
        if (e == "a" or e == "b"):
            #print("{} [{}]->{}  {} {}".format(" "*level, rn, e, pmsg, msg ))
            if msg[0] == e:
                return msg[1:], True
            else:
                return msg, False
            
        sub_result = True
        rest = msg
        for r in e:
            ml = len(msg)
            rl = len(rest)
            matched = pmsg+msg[0:ml-rl]
            #print("{} [{}]->{}  m:{} r:{}".format(" "*level, rn, r, matched, rest ))
            rest, valid = E(matched, rest, rules[r], rules, level+1, r)         
            sub_result = sub_result and valid
            if not sub_result:
                break
        if sub_result:
            return rest, True
    return msg, False

def start(msg, rules):
    return E("", msg, rules["0"], rules, level=0)
```

```python
def match(string, stack):
    if len(stack) > len(string):
        return False
    elif len(stack) == 0 or len(string) == 0:
        return len(stack) == 0 and len(string) == 0
    
    c = stack.pop()
    if isinstance(c, str):
        if string[0] == c:
            return match(string[1:], stack.copy())
    else:
        for rule in rules[c]:
            if match(string, stack + list(reversed(rule))):
                return True
    return False


def valid_messages():
    total = 0
    for message in messages:
        if match(message, list(reversed(rules[0][0]))):
            total += 1
    return total


with open('i.txt') as file:
    rules_raw, messages = tuple(map(lambda x: x.splitlines(), file.read().split('\n\n')))

rules = {}
for rule_raw in rules_raw:
    num, contents = rule_raw.split(': ')
    if contents[0] == '"':
        rules[int(num)] = contents[1]
    else:
        rules[int(num)] = list(map(lambda x: list(map(int, x.split(' '))), contents.split(' | ')))

print("Part 1:", valid_messages())
rules[8] = [[42], [42, 8]]
rules[11] = [[42, 31], [42, 11, 31]]
print("Part 2:", valid_messages())
```
