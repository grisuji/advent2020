```python
import numpy as np
import re

def read_data(file):
    classes={}
    nearby=[]
    myticket = []
    pool = set()
    section = 1
    with open(file,'r') as f:
        for l in f:
            ll = l[:-1]
            if ll == "":
                continue
            if ll == "your ticket:":
                section = 2
                continue
            if ll == "nearby tickets:":
                section = 3
                continue
            if section == 1:
                name = re.findall("[ \w]+:",ll)[0]
                classes[name] = [int(i) for i in re.findall("\d+",ll)]
                pool.add(name)
            elif section == 2:
                myticket = [int(i) for i in re.findall("\d+",ll)]
                
            elif section == 3:
                n = [int(i) for i in re.findall("\d+",ll)]
                nearby.append(n)
                continue
            
    return classes, myticket, nearby, pool

tclasses, tmy, tnear, tnames = read_data("t.txt")
iclasses, imy, inear, inames = read_data("i.txt")
sclasses, smy, snear, snames = read_data("tt.txt")
```

```python
def check_ticket(classes, ticket, pool):
    wrong_l = []
    correct_l =[]
    unregular = []
    for i,n in enumerate(ticket):
        pos = i+1
        #print("-------{}".format(pos))
        found = False
        for name, c in classes.items():
            if (c[0] <= n <= c[1]) or (c[2] <= n <= c[3]):
                correct_l.append(n)
                found = True
            else:
                unregular.append((pos ,name))
        if not found:
            wrong_l.append(n)
        
    if not wrong_l:
        #print(unregular)
        for u in unregular:
            #print(u)
            pool[u[0]].discard(u[1])
    return wrong_l

def sum_unvalid(classes, tickets):
    unvalid=[]
    for t in tickets:
        unvalid += check_ticket(classes, t)
    print(unvalid)
    return sum(unvalid)

def pool(names):
    return { i+1: set(names) for i in range(len(names)) }
    
def clean_pool(pool):
    for i in range(100):
        e = []
        finish = True
        for i, s in pool.items():
            if len(s) == 1:
                el, *_ = s
                e.append(el)
        for ee in e:
            for i, s in pool.items():
                if len(s) != 1:
                    finish = False
                    s.discard(ee)
        if finish:
            return pool
    return {}

```

```python
spool = pool(snames)
for t in snear:
    check_ticket(sclasses, t, spool)
clean_pool(spool)
```
