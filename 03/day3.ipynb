```python
with open("i.txt", 'r') as f:
    grid = [[ c for c in line[:-1] ] for line in f ]
```

```python
def make_step(x,y,l,s,slopex=1, slopey=3):
    return x+slopex, (y+slopey)%l, s+1

def baum(c):
    return c=="#"

def get_field(grid,x,y):
    return grid[x][y]
```

```python
def count_tannen(slopex, slopey):
    x = y = step = 0
    w = len(grid[0])
    h = len(grid)
    tannen = 0
    while 1:
        #print(step, h)
        if baum(get_field(grid, x, y)):
            tannen += 1
        x, y, step = make_step(x, y, w, step, slopex, slopey)
        if x >= h:
            break
    return tannen
count_tannen(1,3)
```

```python
slopes = [(1,1),(3,1),(5,1),(7,1),(1,2)]
result=1
for slopey, slopex in slopes:
    t = count_tannen(slopex, slopey)
    result = result * t
    print(slopex , slopey, t)

result
```
