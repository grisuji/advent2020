```python
def read_data(file):
    d = []
    with open(file,'r') as f:
        for l in f:
            d.append(int(l))
    return d
data = read_data("t.txt")
data
```

```python
data=read_data("t.txt")
def check(data, d):
    sums = set()
    for i in range(len(data)):
        for j in range(i+1, len(data)):
            if data[i] != data[j]:
                sums.add(data[i]+data[j])
    return d in sums
idx = 5
check(data[idx-5:idx], data[idx])
```

```python
for i in range(5, len(data)):
    print("{} {} {}".format(i, data[i], check(data[i-5:i], data[i])))
i=5
data[i-5:i], data[i]
```

```python
data=read_data("i.txt")
pre = 25
for i in range(pre, len(data)):
    c = check(data[i-pre:i], data[i])
    if not c:
        print("{} {}".format(data[i], c))
        break
    
```
