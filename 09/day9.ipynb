```python
def read_data(file):
    with open(file,'r') as f:
        d = [int(l) for l in f ]
    return d
tdata=read_data("t.txt")
pdata=read_data("i.txt")
```

```python
def check(data, d):
    ss = set(data)
    for s in ss:
        if s >= d:
            continue
        delta = d-s
        #print("{}) {} {}".format(d, s, delta))
        if delta != d and delta in ss:
            return True
    return False

def find_security_gap(data, seq_len=25):
    for i in range(seq_len, len(data)):
        c = check(data[i-seq_len:i], data[i])
        if not c:
            # print("{} {}".format(data[i], c))
            break
    return data[i]
```

```python
find_security_gap(tdata, seq_len=5)
```

```python
x=find_security_gap(pdata)
x
```
