```python
import re

def tokenize(e):
    return re.findall("\d+|[+*]|[\(\)]",e)

def read_eq(filename):
    bag = []
    with open(filename,'r') as f:
        for l in f:
            bag.append(l.strip())
    return bag
teq=read_eq("t.txt")
tteq=read_eq("tt.txt")
ieq=read_eq("i.txt")
```

# Part1

```python
"""
start -> AM
AM -> EQ [ +|* EQ ]*
EQ -> number | ( AM ) 
"""

def EQ(t):
    if t[0] == "(":
        result, rest = AM(t[1:])
        if rest[0] == ")":
            return result, rest[1:]
        else:
            print("error {}".format(t))
    
    return int(t[0]), t[1:]
    
def AM(t):
    result, rest = EQ(t)
    while len(rest) > 0:
        if rest[0] == "*":
            res, rest = EQ(rest[1:])
            result *= res
        elif rest[0] == "+":
            res, rest = EQ(rest[1:])
            result += res
        else:
            break
    return result, rest

def solve(e):
    t = tokenize(e)
    res, _ = AM(t)
    return res

[ solve(e) for e in teq]
```

```python
"""
start -> M
M -> A [ * A ]*
A -> E [ + E ]*
E -> number | ( M ) 
"""

def E(t):
    if t[0] == "(":
        result, rest = M(t[1:])
        if rest[0] == ")":
            return result, rest[1:]
        else:
            print("error {}".format(t))
    
    return int(t[0]), t[1:]
    
def M(t):
    result, rest = A(t)
    while len(rest) > 0:
        if rest[0] == "*":
            res, rest = A(rest[1:])
            result *= res
        else:
            break
    return result, rest

def A(t):
    result, rest = E(t)
    while len(rest) > 0:
        if rest[0] == "+":
            res, rest = E(rest[1:])
            result += res
        else:
            break
    return result, rest

def solve2(e):
    t = tokenize(e)
    res, _ = M(t)
    return res
```

```python
tt2 = [ solve2(e) for e in tteq]
i2 = [ solve2(e) for e in ieq]

tt2
```
