```python
import re

def tokenize(e):
    return re.findall("\d+|[+*]|[\(\)]",e)

def read_eq(filename):
    bag = []
    with open(filename,'r') as f:
        for l in f:
            bag.append(l.strip())
    return bag
teq=read_eq("t.txt")
tteq=read_eq("tt.txt")
ieq=read_eq("i.txt")
```

```python
def solve(e):
    t = tokenize(e)
    r, deep = calc(t)
    return r

def calc(t):
    result = 0
    op = "+"
    deep = 0
    for i, c in enumerate(t):
        if deep > 0:
            deep -= 1
            continue
        if c == "(":
             v, deep = calc(t[i+1:])
        elif c == ")":
            return result, i+1
        elif c == "+" or c == "*":
            op = c
            continue
        elif c.isdigit():
            v = int(c)
        if op == "+":
        #    print("{} + {} = {}".format(result, v, result+v))
            result += v
        elif op == "*":
        #    print("{} * {} = {}".format(result, v, result*v))
            result *= v
        else:
            print("error before {}".format(t[i:max(i+1,len(t)-1)]))
    return result, deep    

"""
start -> M
M -> A [ * A ]*
A -> E [ + E ]*
E -> number | ( M ) 
"""

def E(t):
    if t[0] == "(":
        result, rest = M(t[1:])
        if rest[0] == ")":
            return result, rest[1:]
        else:
            print("error {}".format(t))
    
    return int(t[0]), t[1:]
    
def M(t):
    result, rest = A(t)
    while len(rest) > 0:
        if rest[0] == "*":
            res, rest = A(rest[1:])
            result *= res
        else:
            break
    return result, rest

def A(t):
    result, rest = E(t)
    while len(rest) > 0:
        if rest[0] == "+":
            res, rest = E(rest[1:])
            result += res
        else:
            break
    return result, rest

def solve2(e):
    t = tokenize(e)
    res, _ = M(t)
    return res

    
[ solve2(e) for e in tteq]
#solve2(tteq[4])
```
