```python
import numpy as np
def read_data(file):
    
    with open(file,'r') as f:
        start = int(f.readline())
        bus = f.readline()
    busids = [ int(busid) if busid!="x" else 0 for busid in bus[:-1].split(",") ]
    return start, busids 
tstart, tbusids = read_data("t.txt")
istart, ibusids = read_data("i.txt")
```

```python
def part1(start, busids):
    clean = [ bid for bid in busids if bid != 0]
    next = [ bid*(int(start/bid)+1) for bid in clean]
    n = min(next)
    i = next.index(n)
    return (n-start) * clean[i]
```

```python
def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception('modular inverse does not exist')
    else:
        return x % m
    
```

```python
def find_mod(busids):
    p = [ (busid - i)%busid for i, busid in enumerate(busids) if busid!=0]
    return p

def calc_t(busids):
    t = []
    clean = [ bid for bid in busids if bid != 0]
    for i in range(len(clean)):
        copy = clean[:i]+ clean[i+1:]
        t.append(modinv(np.prod(copy), clean[i]))
    return t

# calc product with the expected modulo
def calc_X(busids):
    sum=0
    t = calc_t(busids)
    m = find_mod(busids)
    print(m)
    clean = [ bid for bid in busids if bid != 0]
    for i in range(len(clean)):
        p = np.prod(clean[:i]+ clean[i+1:])
        sum += (t[i]*p*m[i])
    return sum
        
def LCM(busids):
    clean = [ bid for bid in busids if bid != 0]
    return np.prod(clean)
```
