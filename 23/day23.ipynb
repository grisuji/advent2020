```python
class Cup:
    def __init__(self, no):
        self.no = no
        self.left = None
        self.right = None


class Circle:
    def __init__(self, i23):
        self.circle = {}
        last = None
        for no in i23:
            cup = Cup(no)
            if last is not None:
                cup.left = last
                last.right = cup
            else:
                self.cur = cup
            self.circle[no] = cup
            last = cup
        else:
            last.right = self.cur
            self.cur.left = last

        self.max = max(i23)
            
    def move(self):
        cup1 = self.cur.right
        cup2 = cup1.right
        cup3 = cup2.right
        cup4 = cup3.right
        self.cur.right = cup4
        cup4.left = self.cur
        
        dest = self.cur.no-1 or self.max
        while dest in [cup1.no, cup2.no, cup3.no]:
            dest = dest-1 or self.max
        dest_cup = self.circle[dest]
        
        mem = dest_cup.right
        
        dest_cup.right = cup1
        cup1.left = dest_cup
        
        mem.left = cup3
        cup3.right = mem
        
        self.cur = self.cur.right
                
    def play(self, m):
        for i in range(m):
            if i%1_000_000 == 0:
                print(i)
            self.move()
                
    def print_state(self):
        cup = self.cur
        while cup.right != self.cur:
            print("{}".format(cup.no), end='')
            cup = cup.right
        print("")
        
    def get_part2(self):
        cup0 = self.circle[1]
        cup1 = cup0.right
        cup2 = cup1.right
        print("{} {}".format(cup1.no, cup2.no))
        return cup1.no * cup2.no
    
```

```python
my_input=i23
def part2():
    my_nodes = {}

    last_node = None
    for i in my_input:
        curr_node = LinkedListNode(i)
        my_nodes[i] = curr_node

        if last_node is not None:
            last_node.right = curr_node
            curr_node.left = last_node

        last_node = curr_node
    for i in range(len(my_input)+1, 1_000_001):
        curr_node = LinkedListNode(i)
        my_nodes[i] = curr_node
        if last_node is not None:
            last_node.right = curr_node
            curr_node.left = last_node
        last_node = curr_node

    # Complete the circle
    ptr = my_nodes[my_input[0]]
    last_node.right = ptr
    ptr.left = last_node

    ptr = my_nodes[my_input[0]]
    for i in range(10000000):
        p_val = ptr.item

        cup1 = ptr.right
        cup2 = cup1.right
        cup3 = cup2.right

        ptr.right = cup3.right
        ptr.right.left = ptr

        d_val = p_val - 1 or 1000000
        while d_val in (cup1.item, cup2.item, cup3.item):
            d_val = d_val - 1 or 1000000

        d_node = my_nodes[d_val]

        cup3.right = d_node.right
        cup3.right.left = cup3
        d_node.right = cup1
        cup1.left = d_node

        ptr = ptr.right

    while ptr.item != 1:
        ptr = ptr.right

    return ptr.right.item * ptr.right.right.item
print(f"Part 2 solution: {part2()}")
```
