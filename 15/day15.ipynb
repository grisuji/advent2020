```python
tinit=[0,3,6]
iinit=[0,8,15,2,12,1,4]
```

```python
def add(mem, n, turn):
    if n in mem:
        recent = mem[n]
        mem[n] = turn
        return turn - recent
    else:
        mem[n] = turn
        return 0
        
```

```python
def play(init, rounds, verbose=False):
    tmem = {}
    # init
    for turn, last in enumerate(init):
        spoken = add(tmem, last, turn+1) 
        if verbose:
            print("{} {}".format(turn+1,last))
    for turn in range(len(init), rounds):
        last = spoken
        if verbose:
            print("{} {}".format(turn+1,spoken))
        spoken = add(tmem, spoken, turn+1)
    return last
```

```python
play(tinit, 10, True)
```

```python
play(iinit,2020,False)
```

```python
play(tinit, 30000000)
```
