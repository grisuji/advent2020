```python
def read_data(file):
    data=[]
    with open(file, 'r') as f:
        for l in f:
            ll = [ 1 if d == "#" else 0 for d in l[:-1] ]
            data.append(ll)
    return [data]
tdata=read_data("t.txt")
tdata
```

```python
def get_item(data, x,y,z):
    xl = len(data)-1
    yl = len(data[0])-1
    zl = len(data[0][0])-1
    if (0 <= x <= xl) and (0 <= y <= yl) and (0<= z <= zl):
        return data[x][y][z]
    return 0 
get_item(tdata,0,1,2)

def print_slice(data, z, cycle):
    slice=data[z]
    print("z={}".format(z-cycle))
    for x in range(len(slice)):
        row = slice[x]
        s = [ "#" if row[y] == 1 else "." for y in range(len(row)) ]
        print("".join(s))

def count_active_neighbours(data, x,y,z):
    n = [-1, 0, 1]
    sum = 0
    for xx in n:
        for yy in n:
            for zz in n:
                if xx==0 and yy==0 and zz==0:
                    continue
                sum += get_item(data, x+xx, y+yy, z+zz)
    return sum


print_slice(tdata, 0,0)
count_active_neighbours(tdata, 0,1,2)        
```
