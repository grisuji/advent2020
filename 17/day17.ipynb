```python
def read_data(file):
    data=[]
    with open(file, 'r') as f:
        for l in f:
            ll = [ 1 if d == "#" else 0 for d in l[:-1] ]
            data.append(ll)
    return [data]
tdata=read_data("t.txt")
idata=read_data("i.txt")
```

```python
def get_item(data, x,y,z):
    xl = len(data)-1
    yl = len(data[0])-1
    zl = len(data[0][0])-1
    if (0 <= x <= xl) and (0 <= y <= yl) and (0<= z <= zl):
        return data[x][y][z]
    return 0 

def is_active(data, x,y,z):
    return get_item(tdata, x, y, z) == 1


def print_slice(data, z, cycle):
    slice=data[z]
    print("z={}".format(z-cycle))
    for x in range(len(slice)):
        row = slice[x]
        s = [ "#" if row[y] == 1 else "." for y in range(len(row)) ]
        print("".join(s))

def print_cube(data, cycle):
    for s in range(len(data)):
        print_slice(data, s, cycle)
        print(" ")

def count_active_neighbours(data, x,y,z):
    n = [-1, 0, 1]
    sum = 0
    for xx in n:
        for yy in n:
            for zz in n:
                if xx==0 and yy==0 and zz==0:
                    continue
                # print("{} {} {}".format(x+xx,y+yy,z+zz))
                sum += get_item(data, x+xx, y+yy, z+zz)
    return sum

def count_all_active(data):
    sum = 0
    l = len(data)
    for xx in range(l):
        for yy in range(len(data[xx])):
            for zz in range(len(data[xx][yy])):
                sum+=get_item(data,xx,yy,zz)
    return sum

def cycle(data):
    n = [-1, 0, 1]
    lx = len(data)
    ly = len(data[0])
    lz = len(data[0][0])
    cube = []
    for xx in range(-1, lx+1):
        slice = []
        for yy in range(-1, ly+1):
            if lx == 1 and (yy == -1 or yy == ly):
                continue
            row = []
            for zz in range(-1, lz+1):
                if lx == 1 and (zz == -1 or zz == lz):
                    continue
                count = count_active_neighbours(data, xx, yy, zz) 
                #print("{} {} {}".format(xx,yy,zz))
                if is_active(data, xx, yy, zz):
                    c = 1 if count == 3 else 0 #count >= 2 and count <=3 else 0
                else:
                    c = 1 if count >= 3 else 0
                row.append(c)    
            slice.append(row)
        cube.append(slice)
    return cube

def boot(data, cycles=6):
    cube = data
    print(count_all_active(cube))

    for i in range(cycles):
        cube = cycle(cube)
        print(count_all_active(cube))
    return count_all_active(cube)
    

#count_active_neighbours(tdata, 1,1,1)
c1=cycle(tdata)
#print_cube(c1, 1)
```
