```python
def read_data(file):
    data=set()
    with open(file, 'r') as f:
        row=0
        for l in f:
            for col, c in enumerate(l[:-1]):
                if c=="#":
                    data.add((row, col, 0, 0))
            row += 1
            
    return data
tdata=read_data("t.txt")
idata=read_data("i.txt")
tdata
```

```python
def get_item(data, x,y,z,a):
    if (x,y,z,a) in data:
        return 1
    return 0

def get_dims(data):
    x=0
    n=100
    for d in data:
        x=max(x, max(d))
        n=min(n,min(d))
    print(x)
    return n, x

def count_active_neighbours(data, x,y,z,a):
    n = [-1, 0, 1]
    sum = 0
    for xx in n:
        for yy in n:
            for zz in n:
                for aa in n:
                    if xx==0 and yy==0 and zz==0 and aa==0:
                        continue
                    sum += get_item(data, x+xx, y+yy, z+zz, a+aa)
    return sum

def cycle(data):
    cube = set()
    N,X = get_dims(data)
    #print("{} {}".format(N,X))
    for x in range(N-1, X+2):
        for y in range(N-1, X+2):
            for z in range(N-1, X+2):
                for a in range(N-1, X+2):
                    act = get_item(data, x,y,z,a)
                    if act==1 and count_active_neighbours(data,x,y,z,a) in [2,3]:
                        cube.add((x,y,z,a))
                    elif act==0 and count_active_neighbours(data,x,y,z,a) == 3:
                        cube.add((x,y,z,a))
    return cube
                        
def boot(data):
    cube = data
    for c in range(6):
        cube = cycle(cube)
        print(len(cube))
        print(cube)
    return len(cube)
```
