```python
from itertools import product

def read_data(file, dim=3):
    data=set()
    with open(file, 'r') as f:
        row=0
        for l in f:
            for col, c in enumerate(l[:-1]):
                tmp = [0]*dim
                if c=="#":
                    tmp[0]=row
                    tmp[1]=col
                    data.add(tuple(tmp))
            row += 1
            
    return data
t3data=read_data("t.txt", dim=3)
i3data=read_data("i.txt", dim=3)
t4data=read_data("t.txt", dim=4)
i4data=read_data("i.txt", dim=4)
```

```python
def get_item(data, x):
    if x in data:
        return 1
    return 0

def get_dims(data, pos=None):
    x=0
    n=10000
    for d in data:
        if pos is None:
            x=max(x, max(d))
            n=min(n,min(d))
        else:
            x=max(x, d[pos])
            n=min(n, d[pos])
    return n, x

def count_active_neighbours(data, x):
    dim = len(x)
    n = [-1, 0, 1]
    origin = tuple([0]*dim)
    s = 0
    for xx in product(*([n]*dim)):
        if xx == origin: 
            continue
        s += get_item(data, tuple(map(sum, zip(x, xx))) )
    return s

def cycle(data):
    dim = len(next(iter(data)))
    cube = set()
    N,X = get_dims(data)
    n = [*range(N-1, X+2)]
    for xx in product(*([n]*dim)):
        act = get_item(data, xx)
        if act==1 and count_active_neighbours(data,xx) in [2,3]:
            cube.add(xx)
        elif act==0 and count_active_neighbours(data,xx) == 3:
            cube.add(xx)
    return cube
                        
def print_slice(hypercube, dim_tail):
    dim = len(next(iter(hypercube)))
    N0,X0 = get_dims(hypercube, pos=0)
    N1,X1 = get_dims(hypercube, pos=1)
    for x in [*range(N0, X0+1)]:
        print ("".join(["#" if get_item(hypercube, (x,y)+dim_tail)==1 else "." for y in [*range(N1, X1+1)]]))           

    
def boot(data):
    cube = data
    for c in range(6):
        cube = cycle(cube)
    return len(cube)
```
