```python
def read_deck(filename):
    with open (filename, 'r') as f:
        game = {}
        act = "unknown"
        for l in f:
            line = l.strip()
            if not line:
                continue
            if line == "Player 1:" :
                actor = "player1"
                game[actor] = []
            elif line == "Player 2:":
                actor = "player2"
                game[actor] = []
            else:
                game[actor].append(int(line))
    return game
#tdeck = read_deck("t.txt")
ideck = read_deck("i.txt")
```

```python
def round(deck):
    p1 = deck["player1"].pop(0)
    p2 = deck["player2"].pop(0)
    if p1 > p2:
        deck["player1"] += [p1,p2]
    else:
        deck["player2"] += [p2,p1]
```

```python
def eval(deck):
    winner = deck["player1"] if deck["player1"] else deck["player2"]
    s = 0
    for i, v in enumerate(winner[::-1]):
        s += (i+1)*v
    return s

def play(deck):
    for i in range(1000):
        round(deck)
        if not (deck["player1"] and deck["player2"]):
            break
    return i 
```
