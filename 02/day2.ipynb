```python
import pandas as pd
import numpy as np
import re
```

```python
i=pd.read_csv("in.txt", header=None, sep=" ")
i
```

```python
def valid1(row):
    l = row[0].split("-")
    min = int(l[0])
    max = int(l[1])
    c = row[1][0]
    count = row[2].count(c)
    return count >= min and count <= max
    
i["result1"] = i.apply(lambda x: valid1(x) , axis=1)
i["result1"].sum()
```

```python
def valid2(row):
    l = row[0].split("-")
    p1 = int(l[0])
    p2 = int(l[1])
    c = row[1][0]
    passwd = row[2]
    return (passwd[p1-1] == c) ^(passwd[p2-1] == c)
    
i["result2"] = i.apply(lambda x: valid2(x) , axis=1)
i["result2"].sum()
```
