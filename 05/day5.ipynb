```python
def half(x, y):
    return int((x + y + 1)/2)
half(0,127)
```

```python
def F(m, x):
    if m+1 == x:
        return m, m
    return m, half(m, x)-1
a, b = F(0,127)
a, b
```

```python
def B(m, x):
    if m+1 == x:
        return x, x
    return half(m, x), x
B(0,63)
```

```python
def R(m, x):
    if m+1 == x:
        return x, x
    return half(m, x), x
R(0,7)
```

```python
def L(m, x):
    if m+1 == x:
        return m, m
    return m, half(m, x)-1
L(4,7)
```

```python
def decode(seat, verbose=False):
    rm, rx = 0, 127
    for n in range(7):
        if seat[n]=='F':
            rm, rx = F(rm, rx)
        elif seat[n]=='B':
            rm, rx = B(rm, rx)
        if verbose:
            print(rm, rx)
    cm, cx = 0, 7
    if verbose:
        print(20*"-")
    for n in range(7, 10):
        if seat[n]=='R':
            cm, cx = R(cm, cx)
        elif seat[n]=='L':
            cm, cx = L(cm, cx)
        if verbose:
            print(cm, cx)

    return rm*8+cm
decode("FBFBBFFRLR")
```

```python
xx = 0 
for seat in seats:
    #print(seat)
    sid = decode(seat)
    if sid > xx:
        xx = sid
xx
```

```python
s = set([decode(seat) for seat in seats])
for i in range(xx):
    if i-1 in s and i not in s and i+1 in s:
        print(i)
        break
    
```
