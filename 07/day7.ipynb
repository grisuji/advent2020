```python
name_regex = re.compile(r"((\w+ ?)+) bags contain")
num_name_regex = re.compile(r"(\d+ (\w+ ?)+) bags?[.,]")

class bag:
    def __init__(self, rule):
        self.name=""
        self.contains = {}
        self.get_name(rule)
        self.get_next(rule)
    
    def get_name(self, text):
        m=name_regex.search(text)
        self.name=m.groups()[0]
        
    def get_next(self, text):
        m=num_name_regex.search(text)
        if m is None:
            return
        tmp=m.groups()[0]
        if tmp:
            l = tmp.split()
            self.contains[" ".join(l[1:])] = int(l[0])
            s = text.split(tmp)
            if len(s) >1:
                self.get_next(s[1])
```

```python
test="light red bags contain 1 bright white bag, 2 muted yellow bags."
def decode(text):
    key = []
    value = {}
    d=-1
    l=[]
    for f in text.split():
        if f.isdigit():
            d = int(f)
            l = []
            continue
        if f[:3] == "bag":
            name = " ".join(l)
            if not key:
                key = name
            elif d > 0:
                value[name] = d
                d = -1
        if d > 0 or not key:
            l.append(f)
    return key, value
decode(test)
```
