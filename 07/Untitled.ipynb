```python
import re
```

```python
name_regex = re.compile(r"((\w+ ?)+) bags contain")
num_name_regex = re.compile(r"(\d+ (\w+ ?)+) bags?[.,]")

class bag:
    def __init__(self, rule):
        self.name=""
        self.contains = {}
        self.get_name(rule)
        self.get_next(rule)
    
    def get_name(self, text):
        m=name_regex.search(text)
        self.name=m.groups()[0]
        
    def get_next(self, text):
        m=num_name_regex.search(text)
        if m is None:
            return
        tmp=m.groups()[0]
        if tmp:
            l = tmp.split()
            self.contains[" ".join(l[1:])] = int(l[0])
            s = text.split(tmp)
            if len(s) >1:
                self.get_next(s[1])
```
