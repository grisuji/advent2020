```python
def data(file):
    cmds = []
    with open(file,'r') as f:
        for l in f:
            c, n = l.split()
            cmds.append((c,int(n)))
    return cmds
```

```python
def run(cmds, change=-1):
    acc = 0
    i = 0
    markers = {}
    if change == -1:
        change = len(cmds)
    while i < len(cmds):
        cmd, value = cmds[i]
        if markers.get(i,0):
            print("! second call at {}, acc:{}".format(i, acc))
            return 0
        else:
            markers[i]=1
        if i == change:
            cmd = "jmp" if cmd == "nop" else "nop"
        
        if cmd == "acc":
            acc += value
            i += 1
        elif cmd == "jmp":
            i += value
        elif cmd == "nop":
            i += 1
        else:
            print("error at {}".format(i))
            return 0
    print("execute {}, {} at {}, acc:{}".format(cmd, value, i, acc))
    return 1

cmds = data("i.txt")
for j in range(len(cmds)):
    if run(cmds, j):
        break
```
