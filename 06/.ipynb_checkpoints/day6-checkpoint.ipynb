```python
groups = []
group = []
with open("i.txt",'r') as f:
    for l in f:
        if l == "\n":
            groups.append(group)
            group = []
        else:
            group.append(set(l[:-1]))
```

```python
count = 0
for g in groups:
    count += len(set.union(*g))
count
```

```python
count = 0
for g in groups:
    count += len(set.intersection(*g))
count
```
