```python
import re
def read_prg(file):
    with open(file, 'r') as f:
        res = []
        for l in f:
            cmd = l[:-1].split('=')
            if cmd[0][0:4] == "mask":
                res.append(("mask", cmd[1].strip()))
                
            else:
                mem = re.findall(r"mem\[(\d+)\].*",cmd[0])
                res.append((mem[0], int(cmd[1].strip())))
        return res
```

```python
tprog=read_prg("t.txt")
iprog=read_prg("i.txt")
```

```python
def set_bit(value, bit):
    return value | (1<<bit)

def clear_bit(value, bit):
    return value & ~(1<<bit)

def flip_bit(value, bit):
    return value ^ (1<<bit)

def get_mask(m):
    res={}

    for i,c in enumerate(m[::-1]):
        
        if c != "X":
            v = int(c)
            res[i] = v
        else:
            res[i] = -1
            
    return res 

def run_prg(cmds):
    mem = {}
    mask = ""
    for cmd in cmds:
        if cmd[0] == "mask":
            mask = cmd[1]
            continue
        else: 
            v = cmd[1]
            for bit, c in enumerate(mask[::-1]):
                if c == "1":
                    v = set_bit(v, bit)
                elif c == "0":
                    v = clear_bit(v, bit)
                else:
                    pass
            mem[cmd[0]] = v
    return sum(mem.values())
            
def run2_prg(cmds):
    mem = {}
    mask = {}
    for cmd in cmds:
        tmp = {}
        if cmd[0] == "mask":
            mask = cmd[1]
            continue
        else: 
            v = cmd[1]
            mem_add = int(cmd[0])
            for bit, c in enumerate(mask[::-1]):
                if c == '1':
                     mem_add = set_bit(mem_add, bit)
                else:
                    pass
            tmp[mem_add] = v
            for bit, c in enumerate(mask[::-1]):
                if c == 'X':
                    memlist = list(tmp.keys())
                    for mem_add in memlist:
                        tmp[flip_bit(mem_add, bit)] = v
                else:
                    pass
            
            mem.update(tmp)
    return sum(mem.values())
```

```python
run_prg(iprog)
```
