```python
import re
def read_prg(file):
    with open(file, 'r') as f:
        res = []
        for l in f:
            cmd = l[:-1].split('=')
            if cmd[0][0:4] == "mask":
                res.append(("mask", cmd[1].strip()))
                
            else:
                mem = re.findall(r"mem\[(\d+)\].*",cmd[0])
                res.append((mem[0], int(cmd[1].strip())))
        return res
read_prg("t.txt")
```

```python
tprog=read_prg("t.txt")
iprog=read_prg("i.txt")
```

```python
def set_bit(value, bit):
    return value | (1<<bit)

def clear_bit(value, bit):
    return value & ~(1<<bit)

def get_mask(m):
    res={}

    for i,c in enumerate(m[::-1]):
        
        if c != "X":
            v = int(c)
            res[i] = v
        else:
            res[i] = -1
            
    return res 

def run_prg(cmds):
    mem = {}
    mask = {}
    for cmd in cmds:
        if cmd[0] == "mask":
            mask = get_mask(cmd[1])
        else: 
            v = cmd[1]
            for bit, m in mask.items():
                if m == 1:
                    v = set_bit(v, bit)
                elif m == 0:
                    v = clear_bit(v, bit)
                else:
                    pass
            mem[cmd[0]] = v
    sum = 0 
    for m, v in mem.items():
        sum += v
    return sum
            
def run2_prg(cmds):
    mem = {}
    mask = {}
    for cmd in cmds:
        print(cmd)
        tmp = {}
        if cmd[0] == "mask":
            mask = get_mask(cmd[1])
            continue
        else: 
            v = cmd[1]
            mem_add = cmd[0]
            for bit, m in mask.items():
                if m == 1:
                     mem_add = set_bit(v, bit)
                else:
                    pass
            tmp[mem_add] = v
            for bit, m in mask.items():
                if m == -1:
                    memlist = list(tmp.keys())
                    for mem_add in memlist:
                        tmp[set_bit(mem_add, bit)] = v
                        tmp[clear_bit(mem_add, bit)] = v
                else:
                    pass
            
            mem.update(tmp)
    sum = 0 
    for m, v in mem.items():
        sum += v
    return sum
```
