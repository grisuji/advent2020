```python
import re
def read_prg(file):
    with open(file, 'r') as f:
        res = []
        for l in f:
            cmd = l[:-1].split('=')
            if cmd[0][0:4] == "mask":
                res.append(("mask", cmd[1].strip()))
                
            else:
                mem = re.findall(r"mem\[(\d+)\].*",cmd[0])
                res.append((mem[0], int(cmd[1].strip())))
        return res
read_prg("t.txt")
```

```python
def set_bit(value, bit):
    return value | (1<<bit)

def clear_bit(value, bit):
    return value & ~(1<<bit)

def get_mask(m):
    res={}
    flip = 0
    for i,c in enumerate(m[::-1]):
        
        if c != "X":
            v = int(c)
            res[i] = v
        else:
            flip += 2**i
            
    return res, flip    

def run_prg(cmds):
    mem = {}
    mask = {}
    for cmd in cmds:
        if cmd[0] == "mask":
            mask, _ = get_mask(cmd[1])
        else: 
            v = cmd[1]
            for bit, m in mask.items():
                if m == 1:
                    v = set_bit(v, bit)
                else:
                    v = clear_bit(v, bit)
            mem[cmd[0]] = v
    sum = 0 
    for m, v in mem.items():
        sum += v
    return sum
            
def run_prg(cmds):
    mem = {}
    mask = {}
    for cmd in cmds:
        if cmd[0] == "mask":
            mask = get_mask(cmd[1])
        else: 
            v = cmd[1]
            for bit, m in mask.items():
                if m == 1:
                    v = set_bit(v, bit)
                else:
                    v = clear_bit(v, bit)
            mem[cmd[0]] = v
    sum = 0 
    for m, v in mem.items():
        sum += v
    return sum
```
